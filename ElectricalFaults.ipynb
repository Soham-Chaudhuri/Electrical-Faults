```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('detect_dataset.csv')
df.sample(5)
```

```python
df.drop('Unnamed: 7', axis=1, inplace=True)
df.drop('Unnamed: 8', axis=1, inplace=True)
df.sample(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('Output (S)', axis=1)
y = df['Output (S)']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_scaled = pd.DataFrame(sc.transform(x_train))
x_test_scaled = pd.DataFrame(sc.transform(x_test))
x_train_scaled.sample(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
import xgboost
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# lr=LogisticRegression()
rf=RandomForestClassifier()
gb=GradientBoostingClassifier()
ada=AdaBoostClassifier()
xgb=xgboost.XGBClassifier()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
models={}
models['DecisionTreeClassifier']=dt
# models['LogisticRegression']=lr
models['RandomForestClassifier']=rf
models['GradientBoostingClassifier']=gb
models['AdaBoostClassifier']=ada
models['xgboost']=xgb
models['KNeighborsClassifier']=knn

def fit_and_score(models, x_train, x_test, y_train, y_test):
    np.random.seed(42)
    model_scores = {}
    for name, model in models.items():
        model.fit(x_train, y_train)
        pred = model.predict(x_test)
        print(name, accuracy_score(y_test, pred)*100)
        print('\n')
        print(confusion_matrix(y_test, pred))
        print('\n')
        print(classification_report(y_test, pred))
        print('\n')

```

```python
fit_and_score(models, x_train_scaled, x_test_scaled, y_train, y_test)
```

```python
df = pd.read_csv('classData.csv')
df.sample(5)
```

```python
x = df.drop(['G','C','B','A'],axis=1)
y = df[['G','C','B','A']]

y.sample(5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
sc.fit(x_train)
x_train_scaled = pd.DataFrame(sc.transform(x_train))
x_test_scaled = pd.DataFrame(sc.transform(x_test))
x_train_scaled.sample(5)
```

```python
model = RandomForestClassifier(max_depth=10, min_samples_leaf=1, min_samples_split=2, n_estimators=200)
model.fit(x_train_scaled, y_train)
y_pred = pd.DataFrame(model.predict(x_test_scaled))
y_pred.columns = ['G','C','B','A']
print(accuracy_score(y_test, y_pred))
```

```python
y_pred.sample(5)
```

```python
print(classification_report(y_test, y_pred))
```

```python
model = DecisionTreeClassifier()
model.fit(x_train_scaled, y_train)
pred = model.predict(x_test_scaled)
pred = pd.DataFrame(pred)
pred.columns = ['G','C','B','A']
print(classification_report(y_test, pred))
```

```python
model = xgboost.XGBClassifier()
model.fit(x_train_scaled, y_train)
pred = model.predict(x_test_scaled)
pred = pd.DataFrame(pred)
pred.columns = ['G','C','B','A']
print(classification_report(y_test, pred))
```

```python
model = KNeighborsClassifier()
model.fit(x_train_scaled, y_train)
pred = model.predict(x_test_scaled)
pred = pd.DataFrame(pred)
pred.columns = ['G','C','B','A']
print(classification_report(y_test, pred))
```
